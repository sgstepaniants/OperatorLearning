```python
import math
import random
import numpy as np
import torch
import torch.nn as nn
import torch.nn.utils.parametrize as parametrize
import h5py
from pykeops.torch import LazyTensor, generic_sum
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
import time
from scipy.special import gamma, kv

import sys
sys.path.append("..")
from losses import squared_error, relative_squared_error
from rkhs_functions import RKHSFunction
from constraints import Symmetric, FlipSymmetric

use_cuda = torch.cuda.is_available()
tensor = torch.cuda.FloatTensor if use_cuda else torch.FloatTensor
```

```python
mx_true = 1000
x_true = torch.linspace(0, 1, mx_true)

# Generate true Green's function
L = 1
w = 0

X, Y = torch.meshgrid(x_true, x_true, indexing="ij")
G_true = torch.zeros((mx_true, mx_true))
if np.abs(w) < 1e-15:
    G_true = (X + Y - torch.abs(Y - X)) / 2 - X * Y
else:
    for k in range(1, mx_test):
        p = math.pi*k/L
        G_true += 2/L * torch.sin(p*X) * torch.sin(p*Y) / (p**2 - w**2)
```

```python
# Training parameters
num_train = 500
epochs = 10
batch_size = 100

# Noise level on train data
sigma = 0

# Kernel for RKHS
kernel = "Matern5/2"

# Hyperparameter sweep batches
batch_num = 1
total_batches = 180

# Hyperparameters with which to train Green's functions
mx_list_train = np.arange(50, 501, 50)
kernel_width_list_train = [1e-3, 5e-3, 1e-2, 5e-2, 1e-1, 5e-1, 1.0]

# Hyperparameters with which to test Green's functions
mx_list_test = np.arange(50, 501, 50)
kernel_width_list_test = [1e-6, 5e-6, 1e-5, 5e-5, 1e-4, 5e-4, 1e-3, 5e-3, 1e-2, 5e-2, 1e-1, 5e-1, 1.0]
```

```python
combinations = []
for mx_train in mx_list_train:
    for kernel_width_train in kernel_width_list_train:
        for mx_weight in mx_list_train:
            for kernel_width_weight in kernel_width_list_train:
                combinations.append((mx_train, kernel_width_train, mx_weight, kernel_width_weight))
random.Random(4).shuffle(combinations)

total_combinations = len(combinations)
batch_size = math.ceil(total_combinations / total_batches)

batch_combinations = combinations[(batch_num-1)*batch_size:batch_num*batch_size]
print(len(batch_combinations), flush=True)
```

```python
for combination in batch_combinations:
    mx_train, kernel_width_train, mx_weight, kernel_width_weight = combination
    print(f"Combination ({mx_train}, {kernel_width_train}, {mx_weight}, {kernel_width_weight})")
    
    # 1. Generate train data
    data = h5py.File("../generate_data/poisson_gaussianKLE1D_zerobc.hdf5", "r")
    x_train = data[f"mesh{mx_train}"].attrs["mesh"][0]
    
    fs_train = data[f"mesh{mx_train}"][f"kernelwidth{kernel_width_train }"]["forcings"][:, :num_train].T
    us_train_clean = data[f"mesh{mx_train}"][f"kernelwidth{kernel_width_train }"]["solutions"][:, :num_train].T

    x_train = torch.from_numpy(x_train)
    fs_train = torch.from_numpy(fs_train)
    us_train_clean = torch.from_numpy(us_train_clean)

    # add noise to solutions at a given SNR
    us_train = us_train_clean + sigma * torch.std(us_train_clean) * torch.randn(num_train, mx_train)

    # compute the signal to noise of the solutions on the train data
    #snr = torch.sqrt(torch.mean(torch.sum(us_train**2, dim=0)) / torch.mean(torch.sum((us_train-us_train_clean)**2, dim=0)))
    #print('Signal to Noise Ratio: ' + str(snr))

    fs_train = fs_train.type(tensor)
    us_train = us_train.type(tensor)
    
    
    # 2. Train model
    out_meshes = (x_train, x_train)
    ind_divisor = 0
    f_delta = torch.prod(torch.tensor([torch.mean(torch.diff(x)) for x in out_meshes[:ind_divisor+1]]))
    u_delta = torch.prod(torch.tensor([torch.mean(torch.diff(x)) for x in out_meshes[ind_divisor+1:]]))

    x_weight = torch.linspace(x_train.min(), x_train.max(), mx_weight)
    weight_meshes = (x_weight, x_weight)
    weight_delta = torch.prod(torch.tensor([torch.mean(torch.diff(x)) for x in weight_meshes]))
    
    greens_function = RKHSFunction(out_meshes, weight_meshes, kernel=kernel, sigma=kernel_width_weight, dtype=tensor)

    # training example to plot
    example_ind = 0

    num_weights = math.prod([len(x) for x in weight_meshes])
    ones = torch.ones(num_weights).type(tensor)
    K_frob = torch.sqrt(torch.sum((greens_function.K_norm**2) @ ones)).item()
    weight_prefactor = math.sqrt(num_weights) / (weight_delta * K_frob)
    optimizer = torch.optim.Adam(greens_function.parameters(),
                                 lr = 1 * weight_prefactor,
                                 amsgrad=True)

    # RKHS smoothness penalty
    greens_lmbda = 0
    
    # iterate to find the optimal network parameters
    freq = 10
    epoch = 1
    runs = 0
    while epoch <= epochs:
        perm = torch.randperm(num_train)
        for i in range(0, num_train, batch_size):
            inds = perm[i:i+batch_size]
            fs_batch = fs_train[inds, :]
            us_batch = us_train[inds, :]

            G = greens_function()
            G = torch.reshape(G, (mx_train, mx_train))
            us_hat = fs_batch @ G * f_delta

            loss = squared_error(us_hat, us_batch, agg="mean") * u_delta
            loss += greens_lmbda * greens_function.square_norm()

            optimizer.zero_grad()
            # prevent gradient measurement to accumulate
            loss.backward()

            # calculate gradient in each iteration
            optimizer.step()
        
        us_hat = fs_train @ G * f_delta
        """
        if epoch % freq == 0:
            u = us_train[example_ind, :].cpu()
            u_hat = us_hat[example_ind, :].cpu()
            plt.figure(1)
            plt.title('Training Output Example', fontweight='bold')
            plt.plot(x_train, u, label='$u$')
            plt.plot(x_train, u_hat.detach(), '--', label='$\widehat{u}$')
            plt.xlabel('x')
            plt.legend(loc='upper right')
            plt.show()
        """
        
        # compute train rse over all data
        rse = relative_squared_error(us_hat.cpu(), us_train.cpu(), agg="mean").item()
        if epoch == 1 and rse > 1:
            runs += 1
            if runs % 2 == 0:
                eta = 10**(-runs//2)
            else:
                eta = 5*10**(-(runs+1)//2)
            greens_function = RKHSFunction(out_meshes, weight_meshes, kernel=kernel, sigma=kernel_width_weight, dtype=tensor)
            optimizer = torch.optim.Adam(greens_function.parameters(),
                                 lr = eta * weight_prefactor,
                                 amsgrad=True)
            print(f'First epoch had rse = {rse} > 1, setting learning rate to {eta}')
            continue
        print('Epoch {} Relative Squared Error {}'.format(epoch, rse))
        epoch += 1
    
    greens_function.update_mesh((x_true, x_true))
    G = greens_function()
    G = torch.reshape(G, (mx_true, mx_true))
    plt.pcolormesh(G.cpu().detach())
    plt.colorbar()
    plt.show()
    
    G_rse = relative_squared_error(G.cpu().detach(), G_true, agg="none", dim=(0, 1)).item()
    print(f"Green's Function Error: {G_rse}")
    
    # 3. Test model
    for mx_test in mx_list_test:
        for kernel_width_test in kernel_width_list_test:
            data = h5py.File("../generate_data/poisson_gaussianKLE1D_zerobc.hdf5", "r")
            x_test = data[f"mesh{mx_test}"].attrs["mesh"][0]

            fs_test = data[f"mesh{mx_test}"][f"kernelwidth{kernel_width_test}"]["forcings"][:, num_train:].T
            us_test = data[f"mesh{mx_test}"][f"kernelwidth{kernel_width_test}"]["solutions"][:, num_train:].T

            x_test = torch.from_numpy(x_test)
            fs_test = torch.from_numpy(fs_test).type(tensor)
            us_test = torch.from_numpy(us_test).type(tensor)

            greens_function.update_mesh((x_test, x_test))
            G = greens_function()
            G = torch.reshape(G, (mx_test, mx_test))
            
            out_meshes = (x_test, x_test)
            f_delta = torch.prod(torch.tensor([torch.mean(torch.diff(x)) for x in out_meshes[:ind_divisor+1]]))
            us_test_hat = fs_test @ G * f_delta
            test_rse = relative_squared_error(us_test_hat, us_test, agg="mean").item()
            print(f"Test RSE ({mx_test}, {kernel_width_test}): {test_rse}")
```
