```julia
using Pkg
Pkg.add("BenchmarkTools")
Pkg.add("FFTW")
Pkg.add("LinearAlgebra")
Pkg.add("SparseArrays")
Pkg.add("Plots")
Pkg.add("PyPlot")
Pkg.add("DSP")

using BenchmarkTools
using FFTW
using LinearAlgebra
using SparseArrays
using Plots
using PyPlot
using DSP

axis(xflip=true, yflip=true)
```

```julia
# spectral method for periodic 1D Poisson

N = 100 # N grid points on interval [0, 1]
h = 1/N
L = 1
x = range(0, L, length=N)

# solve Poisson with periodic BCs and gauge condition u_hat(0) = 0
f = x.^2 - x .+ 1/6

f_hat = FFTW.fft(f)
freqs = DSP.fftfreq(N, N-1)

# gauge constraint
g = 0

u_hat = zeros(Complex{Float64}, N)
u_hat[1] = g
u_hat[2:N] = f_hat[2:N] ./ (2*pi*freqs[2:N]).^2
u = real(FFTW.ifft(u_hat))

sol = -x.^4/12 .+ x.^3/6 .- x.^2/12 .+ 1/360
Plots.plot(Vector[x, x, x], Vector[f, u, sol], label=["Forcing", "Solution", "Truth"])
```

```julia
# finite differences for periodic 1D Poisson

K = spdiagm(-1 => -ones(N-1), 0 => 2*ones(N), 1 => -ones(N-1))
K[1, N] = -1
K[N, 1] = -1

function DFT_matrix(N)
    return [exp(-2*pi*im*r*c/N) for r in 0:N-1, c in 0:N-1]
end

F = DFT_matrix(N)
F_inv = 1/N * F'
K_hat = F * K * F_inv

K_hat[1, 1] = 1

# gauge constraint
g = 0

# forcing function
f = x.^2 - x .+ 1/6
f_hat = F * f
f_hat[1] = N*g/h^2 

u_hat = h^2 * (K_hat \ f_hat)
u = real(F_inv * u_hat)

sol = -x.^4/12 .+ x.^3/6 .- x.^2/12 .+ 1/360
Plots.plot(Vector[x, x, x], Vector[f, u, sol], label=["Forcing", "Solution", "Truth"])
```

```julia
# covariance matrix C = tau^{2*alpha-d}(-Laplacian + tau^2*Id)^{-alpha}
#tau = 1
#alpha = 1

# gauge constraint
g = 0

# number of samples
n = 10000

fs = zeros(2*n, N)
us = zeros(2*n, N)
for i in 1:(2*n)
    # sample input gaussian random field
    #lambdas = tau^(2*alpha-1)*(4*pi^2*(1:N).^2 .+ tau^2).^(-alpha)
    #coeffs = lambdas .* randn(N)
    f = sin.(pi*x*(1:N)') * (sqrt(2)*randn(N) ./ (pi*float(1:N))) # Brownian bridge
    
    # use DFT to solve Poisson equation with forcing f
    f_hat = FFTW.fft(f)
    freqs = DSP.fftfreq(N, N-1)
    
    u_hat = zeros(Complex{Float64}, N)
    u_hat[1] = g
    u_hat[2:N] = f_hat[2:N] ./ (2*pi*freqs[2:N]).^2
    u = real(FFTW.ifft(u_hat))
    
    # save input forcing and output solution
    fs[i, :] = f
    us[i, :] = u
end

train_fs = fs[1:n, :]
train_us = us[1:n, :]
test_fs = fs[n+1:end, :]
test_us = us[n+1:end, :]

Plots.plot(x, train_us[1, :])
```

```julia
# Perform Gaussian kernel regression over Hilbert space of input functions

function heaviside(x)
    (sign.(x) + 1) / 2
end

ind = 1
f = test_fs[ind, :]

# compute values of gaussian kernels
sigma = 0.03 # kernel width
#kernels = exp.(-sum((train_fs .- f').^2, dims=2)*h / (2*sigma^2))
kernels = heaviside.(-sum((train_fs .- f').^2, dims=2)*h .+ sigma)

u = kernels' * train_us / sum(kernels)
u = dropdims(u, dims=1)

Plots.plot(Vector[x, x], Vector[u, test_us[ind, :]], label=["Solution", "Truth"])
```
